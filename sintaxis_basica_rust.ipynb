# RUST

Rust es un lenguaje de programación enfocado en seguridad, velocidad y concurrencia. Fue creado para escribir código sin que este sea peligroso para la memoria. Es básicamente un C++ en el que el compilador no te deja cometer errores

## ¿Cuándo usar Rust?

Rust se utiliza cuando se necesitan realizar sistemas críticos donde el rendimiento es esencial. Los proyectos en los que conviene utilizar Rust son:

- Sistemas de alto rendimiento: Se utiliza para programa que debe procesar millones de solicitudes por segundo o manejar grandes volúmenes de datos.
- Desarrollo de Microservicios: Se utiliza para crear microservicios ligeros, debido a que consume una fracción de lo que consumiría un servicio escrito en NodeJS o Python, reduciendo costos en la nube
- Sistemas Operativos: Es muy utilizado en el desarrollo de sistemas operativos, drivers de dispositivos, etc. Esto se debe a que se puede usar funciones de alto nivel pero con un rendimiento superios.

Cargo es la herramienta más importante que utilizan los desarrolladores de Rust. No solo se encarga de gestionar paquetes tal como npm o pip, sino que también es un sistema de construcción de proyectos de Rust y un ejecutor de pruebas.

# Sintaxis Básica de Rust


## 1. Variables y Mutabilidad

En Rust las variables se declaran con `let`, tal como se hace en javascript. Sin embargo, a diferencia de javascript, en donde las variables declaradas con let pueden cambiar su valor, en Rust no pueden cambiar su valor, es decir son inmutables. 
Si se necesita que una variable pueda cambiar su valor se debe utilizar `mut` para declarar que puede mutar, quedando `let mut`.

````rust
// Variable inmutable 
let nombre = "Carlos";
// nombre = "Pedro";  // esto causaría un error porque no se puede reasignar una variable inmutable

// Variable mutable
let mut edad = 25;
edad = 26;  // esto sí funciona porque usamos 'mut'
````

### Shadowing

Rust permite declarar una nueva variable con el mismo nombre que una anterior. A esto se le llama shadowing. No es lo mismo que mutar la variable: básicamente se está creando una variable completamente nueva que sustituye a la anterior. Incluso implementando esto se puede cambiar el dato anterior

````rust
let x = 5;          // x es un entero con valor 5
let x = x + 1;      // nuevo x con valor 6 
let x = "texto";    // ahora es un string. 
````

### Constantes

Las constantes se declaran con `const` y siempre se debe indicar el tipo de dato. A diferencia de `let`, las constantes no permiten `mut` ni shadowing, y su valor debe conocerse en tiempo de compilación, es decir, su valor debe estar especificado en el código.

````rust
const MAX_CONEXIONES: u32 = 100;
const PI: f64 = 3.14159;

// Ejemplo de constante que no funcionaría
// const NOMBRE = obtenerNombreDesdeApi();

// no funciona porque el nombre se tiene que obtener y no está especificado en el código
````

## 2. Tipos de Datos

Rust es un lenguaje con **tipado estático**, lo que significa que el compilador necesita saber el tipo de cada variable en tiempo de compilación. Muchas veces Rust lo infiere automáticamente, pero hay casos donde debes indicarlo de forma explícita.

### Tipos Escalares

Son los tipos que representan un único valor.

| Categoría | Tipo | Descripción | Ejemplo |
|-----------|------|-------------|---------|
| **Enteros con signo** | `i8`, `i16`, `i32`, `i64`, `i128` | Números enteros que pueden ser positivos o negativos. El número indica los bits | `let x: i32 = -42;` |
| **Enteros sin signo** | `u8`, `u16`, `u32`, `u64`, `u128` | Solo números positivos (y cero). | `let edad: u8 = 25;` |
| **Punto flotante** | `f32`, `f64` | Números con decimales.  | `let precio: f64 = 9.99;` |
| **Booleano** | `bool` | Solo puede ser `true` o `false` | `let activo: bool = true;` |
| **Carácter** | `char` | Un solo carácter Unicode. Se declara con comillas simples | `let letra: char = 'A';` |
| **Tamaño de arch.** | `isize`, `usize` | Su tamaño depende de la arquitectura (32 o 64 bits). | `let i: usize = 0;` |

### Tipos Compuestos

Agrupan múltiples valores en un solo tipo.

| Tipo | Descripción | Ejemplo |
|------|-------------|---------|
| **Tupla** | Agrupa valores de diferentes tipos. Tiene tamaño fijo | `let persona: (String, u8) = ("Ana".to_string(), 30);` |
| **Array** | Colección de valores del **mismo tipo** con tamaño fijo | `let nums: [i32; 3] = [1, 2, 3];` |
| **String** | Cadena de texto dinámica almacenada en el heap | `let s = String::from("hola");` |
| **&str** | Referencia a una cadena de texto. A diferencia de String, es inmutable | `let s: &str = "hola";` |

````rust
// Enteros
let entero: i32 = -100;
let positivo: u32 = 100;

// Flotantes
let decimal = 3.14;           // f64 por defecto
let decimal_32: f32 = 2.71;

// Booleano y Carácter
let activo = true;
let emoji: char = '🦀';    

// Tupla: se accede por índice con punto
let persona = ("Carlos", 25, true);
let nombre = persona.0;       // "Carlos"
let edad = persona.1;         // 25

// Array: tamaño fijo, mismo tipo
let notas: [f64; 4] = [8.5, 9.0, 7.5, 10.0];
let primera_nota = notas[0];  // 8.5
````

## 3. Expresiones y Declaraciones

### Expresiones
Las expresiones evaúan algo y devuelven un valor. Se considera expresiones el llamado de funciones, expresiones mátematicas, etc. De hecho, incluso los bloques de código se consideran expresiones 

```Rust
// Expresión matemática
15 + 35 

// Llamado a función
getNombre()

// Bloque de código
{
    let a = 5;
    let b = 6;
    a + b
}
```

Cabe resaltar que las expresiones no terminan con punto y coma


### Declaraciones
Las declaraciones son instrucciones que realizan una acción pero no devuelven un valor. Las declaraciones terminan con punto y coma. Normalmente dentro de las declaraciones se consiguen expresiones

```Rust
// Expresión matemática
let a = 15 + 35; 

// Llamado a función
let nombre = getNombre();

// Bloque de código
let suma = {
    let a = 5;
    let b = 6;
    a + b
};
```

## 4. Funciones

Las funciones se declaran con `fn`. Se deben tipar los parámetros y además, si la función retorna un valor, se indica el tipo de retorno con `->`. Además, como vimos anteriormente, las expresiones ejecutan operaciones y retornan un valor, por lo tanto, normalmente para retornar valores de las funciones se termina con expresiones, es decir, se retorna con una línea sin punto y coma(aunque también puedes usar `return`).

````rust
// Función sin retorno
fn saludar(nombre: &str) {
    println!("Hola, {}!", nombre);
}

// Función con retorno (la última expresión sin punto y coma es el retorno)
fn sumar(a: i32, b: i32) -> i32 {
    a + b   // No lleva ; porque es el valor de retorno
}

// También se puede usar return de forma explícita
fn es_mayor_de_edad(edad: u8) -> bool {
    return edad >= 18;
}

// Función con múltiples retornos usando tupla
fn cociente_y_residuo(a: f64, b: f64) -> (f64, f64) {
    let cociente = a / b;
    let residuo = a % b;
    (cociente, residuo)
}
````

## 6. Condicionales

Los condicionales en Rust funcionan igual que en otros lenguajes, la única diferencia es que la condición no va entre paréntesis.

Por otro lado, como podemos ver el if es un bloque de código, por lo tanto se puede deducir que `if` es una **expresión**, lo que significa que se puede usar para asignar valores.

````rust
let temperatura = 35;

if temperatura > 30 {
    println!("Hace mucho calor");
} else if temperatura > 20 {
    println!("Clima agradable");
} else {
    println!("Hace frío");
}

// if como expresión (es como utilizar un operador ternario en JS)
let mensaje = if temperatura > 30 { "caliente" } else { "fresco" };
println!("Está {}", mensaje);
````

### Match

`match` es el equivalente al `switch` de otros lenguajes. Sin embargo, Rust te obliga a cubrir todos los casos posibles, lo que evita bugs. Se usa `_` como caso por defecto (equivalente al `default`).

````rust
let codigo_http = 404;

let estado = match codigo_http {
    200 => "OK",
    301 => "Redirección",
    404 => "No encontrado",
    500 => "Error del servidor",
    _   => "Código desconocido",  // _ cubre todos los demás casos
};

println!("Estado: {}", estado);

// Match con rangos
let nota = 85;
let calificacion = match nota {
    90..=100 => "A",
    80..=89  => "B",
    70..=79  => "C",
    _        => "F",
};
````

## 7. Bucles

Rust tiene tres tipos de bucles: `loop`, `while` y `for`. Además se mantienen las mismas palabras reservadas como `break`, `continue`

### loop

Es un bucle infinito. Se ejecuta hasta que explícitamente lo detengas con `break`. Puede devolver un valor al romper el ciclo.

````rust
let mut contador = 0;

// loop básico
loop {
    contador += 1;
    if contador == 5 {
        break;  // Sale del bucle
    }
}

// loop que retorna un valor
let mut num = 0;
let resultado = loop {
    num += 1;
    if num * num > 100 {
        break num;  // Retorna el valor de 'num' al salir
    }
};
````

### while

Se ejecuta mientras una condición sea verdadera. Igual que en otros lenguajes, pero sin paréntesis en la condición.

````rust
let mut intentos = 3;

while intentos > 0 {
    println!("Te quedan {} intentos", intentos);
    intentos -= 1;
}
````

### for

Este es el bucle más usado. Itera sobre cualquier cosa que implemente el trait `Iterator`.

````rust
// Iterar sobre un rango
for i in 0..5 {
    println!("Número: {}", i);  // Imprime 0, 1, 2, 3, 4
}

// Rango inclusivo 
for i in 0..=5 {
    println!("Número: {}", i);  // Imprime 0, 1, 2, 3, 4, 5
}

// Iterar sobre un array
let frutas = ["manzana", "banana", "cereza"];
for fruta in &frutas {
    println!("Fruta: {}", fruta);
}

// Iterar con índice usando enumerate
for (i, fruta) in frutas.iter().enumerate() {
    println!("{}: {}", i, fruta);
}
````

## 8. Ownership (Propiedad)

Este es el concepto más importante de Rust y es lo que lo diferencia de todos los demás lenguajes. El ownership es el sistema que usa Rust para manejar la memoria sin necesidad de un garbage collector (como Java o Go) y sin que el programador tenga que liberar memoria manualmente (como en C/C++). 

Antes de entender el ownership debemos entender primero el Stack y el Heap
### Stack y Heap:
- El stack es una memoria rápida y ordenada donde se almacenan datos de tamaño fijo y conocido (como un entero o un float), su gestión es automática.
- El heap es una memoria más flexible y grande donde se guardan datos de tamaño dinámico (como un vector o un String); requiere gestión manual y es más lenta.

### Las 3 Reglas del Ownership

1. Cada valor en Rust tiene una única variable que es su propietaria (owner)
2. Solo puede haber un propietario a la vez
3. Cuando el propietario sale de su ámbito (scope), el valor se libera automáticamente de la memoria

````rust
// Ejemplo básico de ownership
{
    let mensaje = String::from("hola");  // la variable mensaje es el propietario del String
    println!("{}", mensaje);             // Se puede usar normalmente
}   // Aquí la variable mensaje sale del scope y la memoria se libera automáticamente
````

### Move (Transferencia de propiedad)

Cuando se asigna un valor de un tipo complejo (como `String`) a otra variable, Rust no lo copia, sino que **mueve** la propiedad. La variable original deja de ser válida.

````rust
let nombre = String::from("Carlos");
let otro_nombre = nombre;  // La propiedad se mueve a otro_nombre

// println!("{}", nombre);     // Esto daría error ya que nombre ya no es válido
println!("{}", otro_nombre);   // Esto sí funciona
````

> NOTA: Los tipos simples como enteros, flotantes y booleanos **sí se copian** automáticamente porque son baratos de copiar. Este comportamiento solo aplica a tipos que se almacenan en el heap (como `String`, vectores, etc.).

````rust
// Los tipos simples se copian, no se mueven
let a = 5;
let b = a;   // Se copia el valor
println!("a = {}, b = {}", a, b);  // Ambos son válidos

// Para copiar un string se utiliza .clone()
let original = String::from("hola");
let copia = original.clone();  // Se crea una copia independiente en memoria
println!("original = {}, copia = {}", original, copia);  // Ambos funcionan
````

### Ownership en Funciones

Pasar un valor a una función funciona igual que asignarlo a otra variable: se **mueve** la propiedad. Esto significa que después de pasar un `String` a una función, ya no puedes usar esa variable.

````rust
fn imprimir(texto: String) {
    println!("{}", texto);
}   // 'texto' sale del scope y se libera la memoria

let saludo = String::from("Hola mundo");
imprimir(saludo);            // La propiedad se mueve a la función
// println!("{}", saludo);  // Esto daría error ya que saludo pertenece a la función imprimir y ya no es válido aquí
````

### Referencias y Borrowing (Préstamo)

Para no perder la propiedad cada vez que pasas un valor a una función, Rust permite prestar el valor usando referencias con `&`. Básicamente se pasa una referencia del propietario a la función.

````rust
// Referencia inmutable: solo puede leer, no modificar
fn calcular_longitud(texto: &String) -> usize {
    texto.len()
}   // la variable texto es solo una referencia, así que NO se libera nada

let mensaje = String::from("Hola Mundo");
let longitud = calcular_longitud(&mensaje);  // Prestamos el valor con &
println!("'{}' tiene {} caracteres", mensaje, longitud);  // mensaje sigue siendo válido
````

````rust
// Referencia mutable: permite modificar el valor prestado
fn agregar_exclamacion(texto: &mut String) {
    texto.push_str("!!!");
}

let mut frase = String::from("Te odio ujap");
agregar_exclamacion(&mut frase);   // Prestamos de forma mutable con &mut
println!("{}", frase);             // Imprime: "Te odio ujap!!!"
````

### Reglas del Borrowing

El compilador de Rust es muy estricto con las referencias para evitar problemas de concurrencia:

- Puedes tener muchas referencias inmutables (`&T`) al mismo tiempo
- Solo puedes tener una referencia mutable (`&mut T`) a la vez
- **No puedes** mezclar referencias mutables e inmutables al mismo dato al mismo tiempo

````rust
let mut dato = String::from("hola");

// Esto es válido: múltiples referencias inmutables
let r1 = &dato;
let r2 = &dato;
println!("{} y {}", r1, r2);

// Esto es válido: una sola referencia mutable
let r3 = &mut dato;
r3.push_str(" mundo");

// esto no es válido: mezclar referencia inmutable y mutable
// let r4 = &dato;
// let r5 = &mut dato;  // esto daría error debido a que no puedes tener &mut mientras exista &
````


## 10. Manejo de Errores (Error Handling)

En la mayoría de lenguajes (Java, Python, JavaScript), los errores se manejan con excepciones. El problema es que las excepciones son invisibles: cualquier función puede lanzar una excepción en cualquier momento y si no la atrapas, tu programa se cae.

Rust toma un enfoque diferente: no tiene excepciones. En su lugar, te obliga a manejar los errores de forma explícita usando los tipos `Result` y `Option`. Esto significa que el compilador **no te deja ignorar** un posible error; si una función puede fallar, debes decidir qué hacer con ese caso.

### Result\<T, E\>

`Result` se usa para operaciones que **pueden fallar**. Es un enum con dos variantes:

- `Ok(valor)` → la operación salió bien. `valor` es del tipo `T` (el tipo exitoso)
- `Err(error)` → la operación falló. `error` es del tipo `E` (el tipo del error)

```rust
// Definición interna de Result (ya viene incluido en Rust)
enum Result<T, E> {
    Ok(T),    // Éxito: contiene el valor de tipo T
    Err(E),   // Error: contiene el error de tipo E
}
```

**Comparación con try/catch:**
| Otros lenguajes (try/catch) | Rust (Result) |
|---|---|
| El error es invisible, puede explotar en cualquier momento | El error es parte del tipo de retorno, siempre visible |
| Si olvidas el try/catch, el programa se cae | Si no manejas el Result, el compilador te da un error |
| No sabes qué errores puede lanzar una función | Sabes exactamente qué tipo de error puede devolver |


````rust
// Función que retorna Result: f64 si sale bien, String si falla
fn dividir(a: f64, b: f64) -> Result<f64, String> {
    if b == 0.0 {
        Err(String::from("No se puede dividir por cero"))  // Retorna error con mensaje
    } else {
        Ok(a / b)  // Retorna el resultado envuelto en Ok
    }
}

// Forma 1: Usar match para manejar ambos casos explícitamente
let resultado = dividir(10.0, 0.0);
match resultado {
    Ok(valor) => println!("Resultado: {}", valor),
    Err(e) => println!("Error: {}", e),
}

// Forma 2: unwrap() - extrae el valor de Ok, pero si es Err el programa se cae (panic)
// Útil para pruebas rápidas, NO recomendado en producción
let valor = dividir(10.0, 2.0).unwrap();
println!("Con unwrap: {}", valor);

// Forma 3: unwrap_or() - extrae el valor de Ok, o usa un valor por defecto si es Err
let valor_seguro = dividir(10.0, 0.0).unwrap_or(0.0);
println!("Con unwrap_or: {}", valor_seguro);  // Imprime 0.0 porque falló

// Forma 4: unwrap_or_else() - como unwrap_or pero con un closure para manejar el error
let valor = dividir(10.0, 0.0).unwrap_or_else(|error| {
    println!("Ocurrió un error: {}", error);
    -1.0  // valor por defecto
});
println!("Con unwrap_or_else: {}", valor);
````


### Option\<T\>

En muchos lenguajes existe `null` o `nil` para representar la ausencia de un valor. El problema es que si intentas usar un valor que es `null` sin verificarlo, tu programa se cae (el famoso `NullPointerException`). Rust elimina este problema por completo: **no existe `null`**. En su lugar, usa `Option`.

`Option` se usa cuando un valor **puede o no existir**. Es un enum con dos variantes:

- `Some(valor)` → sí hay un valor
- `None` → no hay valor (es como null, pero seguro porque el compilador te obliga a verificar)

```rust
// Definición interna de Option (ya viene incluido en Rust)
enum Option<T> {
    Some(T),   // Hay un valor de tipo T
    None,      // No hay valor
}
```

**Option vs Result: cuándo usar cada uno**
| Situación | Usa |
|---|---|
| Un valor puede no existir (buscar algo, campo opcional) | `Option<T>` |
| Una operación puede fallar y necesitas saber por qué | `Result<T, E>` |

La diferencia es que `Option` solo dice hay o no hay, mientras que `Result` además te dice **por qué** falló (el `Err` contiene la razón del error).


````rust
// Ejemplo básico con match
let nombre: Option<String> = Some(String::from("Carlos"));
match nombre {
    Some(n) => println!("Hola, {}", n),
    None => println!("No hay nombre"),
}

// Ejemplo práctico: acceder a un elemento de un array
// .get() retorna Option: Some si el índice existe, None si está fuera de rango
let frutas = ["manzana", "banana", "cereza"];

let primera = frutas.get(0);    // Some("manzana")
let fuera = frutas.get(100);    // None (no existe el índice 100)

println!("Primera fruta: {:?}", primera);
println!("Índice 100: {:?}", fuera);

// Métodos útiles de Option:

// unwrap_or: extrae el valor o usa un valor por defecto
let fruta = frutas.get(100).unwrap_or(&"no encontrada");
println!("Fruta: {}", fruta);  // "no encontrada"

// is_some() / is_none(): verificar si hay o no valor
let valor: Option<i32> = Some(42);
println!("Tiene valor: {}", valor.is_some());   // true
println!("Está vacío: {}", valor.is_none());    // false

// map: transformar el valor dentro del Option (si existe)
let numero: Option<i32> = Some(5);
let doble = numero.map(|n| n * 2);  // Some(10)
println!("Doble: {:?}", doble);

let vacio: Option<i32> = None;
let doble_vacio = vacio.map(|n| n * 2);  // None (no hay nada que transformar)
println!("Doble de None: {:?}", doble_vacio);
````

### Operador `?`

Cuando tienes varias operaciones que pueden fallar seguidas, escribir `match` para cada una se vuelve muy repetitivo. El operador `?` es **azúcar sintáctica** que simplifica esto: si el valor es `Ok` extrae el valor y continúa, si es `Err` retorna el error inmediatamente.

Básicamente, esto:
```rust
let valor = match operacion() {
    Ok(v) => v,           // Si salió bien, usa el valor
    Err(e) => return Err(e),  // Si falló, retorna el error
};
```

Es equivalente a esto con `?`:
```rust
let valor = operacion()?;  // Hace exactamente lo mismo pero en una línea
```

> **NOTA:** El operador `?` solo se puede usar dentro de funciones que retornan `Result` o `Option`, ya que necesita un tipo al cual propagar el error.


````rust
// Ejemplo: función que encadena varias operaciones que pueden fallar
fn obtener_longitud(texto: Option<&str>) -> Result<usize, String> {
    // texto.ok_or() convierte un Option en un Result:
    //   Some(valor) -> Ok(valor)
    //   None        -> Err("No hay texto")
    // El ? al final hace: si es Ok, extrae el valor. Si es Err, retorna el error.
    let t = texto.ok_or(String::from("No hay texto"))?;
    Ok(t.len())
}

// Caso exitoso: el Option tiene un valor
let resultado = obtener_longitud(Some("Hola Rust"));
println!("Longitud: {:?}", resultado);  // Ok(9)

// Caso fallido: el Option es None, así que ? propaga el error
let error = obtener_longitud(None);
println!("Error: {:?}", error);  // Err("No hay texto")

// Ejemplo más completo: encadenar múltiples operaciones con ?
fn procesar_numero(texto: Option<&str>) -> Result<i32, String> {
    let t = texto.ok_or(String::from("No hay texto"))?;           // Paso 1: verificar que existe
    let num = t.parse::<i32>().map_err(|e| e.to_string())?;       // Paso 2: convertir a número
    Ok(num * 2)                                                     // Paso 3: duplicar
    // Sin el operador ? esto serían 2 bloques match anidados
}

println!("{:?}", procesar_numero(Some("21")));    // Ok(42)
println!("{:?}", procesar_numero(Some("abc")));   // Err: no se puede parsear
println!("{:?}", procesar_numero(None));           // Err: "No hay texto"
````

## 11. Introducción a la Orientación a Objetos en Rust

Rust **no tiene clases** como Java, Python o C++. Sin embargo, permite lograr los mismos conceptos de la programación orientada a objetos usando una combinación de:

- **Structs** → representan los datos (como los atributos de una clase)
- **Bloques `impl`** → definen los métodos asociados a un struct (como los métodos de una clase)
- **Traits** → definen comportamiento compartido (como las interfaces en Java)

| Concepto OOP | En Java/Python | En Rust |
|---|---|---|
| Clase con atributos | `class Persona { String nombre; }` | `struct Persona { nombre: String }` |
| Métodos | Definidos dentro de la clase | Definidos en un bloque `impl` |
| Constructor | `new Persona()` o `__init__` | Función asociada (por convención `::new()`) |
| Interfaces | `interface` / clases abstractas | `trait` |
| Herencia | `extends` / herencia de clases | No existe. Se usa **composición** y **traits** |
| Polimorfismo | Métodos virtuales / override | Traits + `dyn` (dispatch dinámico) |

### Métodos con `impl` (el equivalente a los métodos de una clase)

En Rust, los métodos se definen en un bloque `impl` (implementation) separado del struct. Hay dos tipos:

- **Métodos de instancia**: reciben `&self` como primer parámetro (equivalente al `this` de Java o `self` de Python). Operan sobre una instancia del struct.
- **Funciones asociadas**: NO reciben `self`. Se llaman con `::` en vez de `.` (como los métodos estáticos en Java). Por convención, se usa `new()` como constructor.

````rust
struct Rectangulo {
    ancho: f64,
    alto: f64,
}

impl Rectangulo {
    // Función asociada (constructor): no recibe &self, se llama con ::
    fn new(ancho: f64, alto: f64) -> Rectangulo {
        Rectangulo { ancho, alto }
    }

    // Método de instancia: recibe &self (referencia inmutable)
    fn area(&self) -> f64 {
        self.ancho * self.alto
    }

    // Método de instancia: recibe &self
    fn perimetro(&self) -> f64 {
        2.0 * (self.ancho + self.alto)
    }

    // Método que modifica el struct: recibe &mut self (referencia mutable)
    fn escalar(&mut self, factor: f64) {
        self.ancho *= factor;
        self.alto *= factor;
    }

    // Método que verifica una condición
    fn es_cuadrado(&self) -> bool {
        self.ancho == self.alto
    }
}

// Usar la función asociada como constructor (similar a new Rectangulo() en Java)
let mut rect = Rectangulo::new(10.0, 5.0);

// Llamar métodos de instancia con punto (igual que en cualquier lenguaje OOP)
println!("Área: {}", rect.area());           // 50.0
println!("Perímetro: {}", rect.perimetro()); // 30.0
println!("¿Es cuadrado? {}", rect.es_cuadrado()); // false

// Llamar método que modifica el struct
rect.escalar(2.0);
println!("Área después de escalar: {}", rect.area()); // 200.0
````

### Traits (el equivalente a las interfaces)

Un **trait** define un conjunto de métodos que un tipo debe implementar. Es similar a una `interface` en Java o un protocolo en Swift. La idea es: "si implementas este trait, garantizas que tienes estos métodos".

Los traits pueden tener:
- **Métodos sin cuerpo**: el struct que lo implemente debe definirlos (como métodos abstractos)
- **Métodos con implementación por defecto**: el struct puede usarlos tal cual o sobreescribirlos

````rust
// Definimos un trait con un método obligatorio y uno con implementación por defecto
trait Describible {
    // Método obligatorio: cada struct DEBE implementarlo
    fn descripcion(&self) -> String;

    // Método con implementación por defecto: es opcional sobreescribirlo
    fn mostrar(&self) {
        println!(">> {}", self.descripcion());
    }
}

// Struct Libro
struct Libro {
    titulo: String,
    autor: String,
    paginas: u32,
}

// Implementamos el trait Describible para Libro
impl Describible for Libro {
    fn descripcion(&self) -> String {
        format!("'{}' por {} ({} páginas)", self.titulo, self.autor, self.paginas)
    }
    // No sobreescribimos mostrar(), así que usa la implementación por defecto
}

// Struct Pelicula
struct Pelicula {
    titulo: String,
    director: String,
    duracion_min: u32,
}

// Implementamos el mismo trait para Pelicula
impl Describible for Pelicula {
    fn descripcion(&self) -> String {
        format!("'{}' dirigida por {} ({}min)", self.titulo, self.director, self.duracion_min)
    }

    // Sobreescribimos mostrar() con una versión personalizada
    fn mostrar(&self) {
        println!("🎬 {}", self.descripcion());
    }
}

let libro = Libro {
    titulo: String::from("El Principito"),
    autor: String::from("Saint-Exupéry"),
    paginas: 96,
};

let pelicula = Pelicula {
    titulo: String::from("Inception"),
    director: String::from("Nolan"),
    duracion_min: 148,
};

libro.mostrar();     // Usa la implementación por defecto: >> 'El Principito' por Saint-Exupéry
pelicula.mostrar();  // Usa la versión sobreescrita: 🎬 'Inception' dirigida por Nolan
````

### Polimorfismo con Traits

En lenguajes como Java, el polimorfismo se logra con herencia: un método recibe un tipo padre y acepta cualquier hijo. En Rust, el polimorfismo se logra con **traits**: una función puede aceptar **cualquier tipo que implemente un trait específico**.

Hay dos formas de hacerlo:

| Forma | Sintaxis | Cuándo usarla |
|---|---|---|
| **Genéricos** (dispatch estático) | `fn mostrar(item: &impl Trait)` | Cuando el compilador conoce el tipo concreto. Más rápido. |
| **Trait objects** (dispatch dinámico) | `fn mostrar(item: &dyn Trait)` | Cuando necesitas una colección de tipos diferentes que comparten un trait. |

````rust
// --- Reutilizamos los structs y el trait Describible definidos arriba ---

// Forma 1: Genéricos con impl (dispatch estático)
// Esta función acepta CUALQUIER tipo que implemente Describible
fn mostrar_info(item: &impl Describible) {
    println!("Info: {}", item.descripcion());
}

// Funciona con Libro y Pelicula porque ambos implementan Describible
mostrar_info(&libro);
mostrar_info(&pelicula);

// Forma 2: Trait objects con dyn (dispatch dinámico)
// Permite crear una colección de tipos DIFERENTES que comparten el mismo trait
let coleccion: Vec<&dyn Describible> = vec![&libro, &pelicula];

println!("\n--- Mi colección ---");
for item in &coleccion {
    item.mostrar();  // Cada uno usa su propia implementación
}
````

### Composición en vez de Herencia

Rust **no tiene herencia de structs**. En lugar de crear una clase padre y extenderla (como en Java), Rust favorece la **composición**: incluir un struct dentro de otro y delegar comportamiento.

```
// En Java harías esto (herencia):
class Animal { String nombre; }
class Perro extends Animal { String raza; }

// En Rust haces esto (composición):
struct Animal { nombre: String }
struct Perro { animal: Animal, raza: String }
```

Esta decisión de diseño evita muchos problemas conocidos de la herencia (como el "diamond problem") y hace el código más flexible.

````rust
// Ejemplo de composición: un Motor es parte de un Auto

struct Motor {
    caballos_de_fuerza: u32,
    tipo_combustible: String,
}

impl Motor {
    fn new(hp: u32, combustible: &str) -> Motor {
        Motor {
            caballos_de_fuerza: hp,
            tipo_combustible: String::from(combustible),
        }
    }

    fn encender(&self) {
        println!("Motor de {} HP encendido ({})", self.caballos_de_fuerza, self.tipo_combustible);
    }
}

struct Auto {
    marca: String,
    motor: Motor,  // Composición: Auto TIENE un Motor (no hereda de Motor)
}

impl Auto {
    fn new(marca: &str, hp: u32, combustible: &str) -> Auto {
        Auto {
            marca: String::from(marca),
            motor: Motor::new(hp, combustible),
        }
    }

    fn arrancar(&self) {
        println!("{} arrancando...", self.marca);
        self.motor.encender();  // Delega al motor
    }
}

let mi_auto = Auto::new("Toyota", 180, "Gasolina");
mi_auto.arrancar();
````
